```python
import random
import math
import seaborn
import PIL
import numpy as np
import soundfile as sf
from matplotlib import rcParams
rcParams['font.family'] = 'monospace' #fix my local problem
import matplotlib.pylab as plt
import scipy.stats as st
from PIL import Image
from math import gcd
from mpl_toolkits.mplot3d.axes3d import Axes3D, get_test_data
from matplotlib import cm
from scipy.stats import multivariate_normal
from scipy.stats import norm
from scipy.io.wavfile import read


#Task 1
def LCGinit(seed):
    global x, a, c, m
    #The GNU C library's rand() in stdlib.h
    a = 1103515245
    c = 12345
    m = 2**31 - 1
    x = seed

def LCG():
    global x
    x = (a*x + c) % m
    return x

LCGinit(123)

#Task 2
def unif(a, b):
    x = a + (b - a) * LCG()
    return x 

numb = 1000

border_a = 0
border_b = 10

LCG_data = []
ddata = []
for i in range(numb):
    x = LCG()
    LCG_data.append(x/m)

LCG_unif = []
for i in range(numb):
    x = LCG()
    x = unif(border_a, border_b)
    LCG_unif.append(x/m*0.1)

print("Task #1 and task #2:")
data = np.vstack([LCG_data, LCG_unif]).T
plt.hist(data, bins=10, alpha=0.7, label=['LCG', 'Uniform'])
plt.legend(loc='upper right')
plt.show()
```

```python
#Task 5

numb = 1000
bins = 100

#N(0,1^2)
norm1 = np.random.normal(0, 1, numb)
plt.hist(norm1, bins, color='blue', alpha=0.7, label = "N(0,1^2)")
plt.legend(loc='upper left')
plt.show()

#N(3,3^2)
norm2 = np.random.normal(3, 3**2, numb)
plt.hist(norm2, bins, color='green', alpha=0.7, label = "N(3,3^2)")
plt.legend(loc='upper left')
plt.show()

#Unif(0,1)
unif1 = np.random.uniform(0, 1, numb)
plt.hist(unif1, bins, color='blue', alpha=0.7, label = "Unif(0,1)")
plt.legend(loc='upper left')
plt.show()


#Unif(-1,1)
unif2 = np.random.uniform(-1, 1, numb)
plt.hist(unif2, bins, color='green', alpha=0.7, label = "Unif(-1,1)")
plt.legend(loc='upper left')
plt.show()


#Gamma(1,2)
gamma = np.random.gamma(1, 2, numb)
plt.hist(gamma, bins, facecolor='blue', alpha=0.7, label = "Gamma(1,2)")
plt.legend(loc='upper left')
plt.show()


#LogN(.08,2^2)
logn = np.random.lognormal(.08, 2, numb)
plt.hist(logn, bins, facecolor='green', alpha=0.7, label = "LogN(.08,2^2)")
plt.legend(loc='upper left')
plt.show()
```

```python
#Task 6

fig = plt.figure(figsize=plt.figaspect(0.5))
plot = fig.add_subplot(1, 2, 1, projection='3d')

x = np.linspace(0, 3, 100)
y = np.linspace(0, 3, 100)
x, y = np.meshgrid(x, y)
z = st.norm.pdf(x, loc=1, scale=0.3)

surf = plot.plot_surface(x, y, z)
plt.show()
```

```python
#Task 7

fig = plt.figure()
plot = fig.gca(projection='3d')

mu_x = 0
var_x = 5

mu_y = 0
var_y = 10

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
x, y = np.meshgrid(x,y)

pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x; pos[:, :, 1] = y
rv = multivariate_normal([mu_x, mu_y], [[var_x, 0], [0, var_y]])

surf = plot.plot_surface(x, y, rv.pdf(pos))
plt.show()
```
