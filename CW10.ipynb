Zad 1.
Policz korelacje: pearson, spearman, kendall dla zbioru: 

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats

f = lambda x: (x)
x_tr = np.linspace(0., 5, 200)
y_tr = f(x_tr)
x = stats.uniform(1,3).rvs(100)
y = f(x) + stats.norm(0,0.1).rvs(len(x))

plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,5])
axes.set_ylim([0,5])
plt.plot(x_tr, y_tr, '--k');
plt.plot(x, y, 'ok', ms=3);
plt.show()

print(stats.pearsonr(x,y))
print(stats.spearmanr(x,y))
print(stats.kendalltau(x,y))
```

 Zad 6.
Proszę wygenerować dane tak by 1. i 3. współrzędna były skorelowane a pozostałe nie. Proszę zwizualizować tą zależność za pomocą

    sns.pairplot
    sns.heatmap


```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats

n=1000
mu = np.array([5.0, 0.0, 2.0, 1.0])

r = np.array([
        [1, 0, 1, 0],
        [0, 1, 0, 0],
        [1, 0, 1, 0],
        [0, 0, 0, 1]
    ])
data = np.random.multivariate_normal(mu, r, size=N)

x1 = data[:,0]
x2 = data[:,1]
x3 = data[:,2]
x4 = data[:,3]

X=np.stack((x1,x2,x3,x4),1)

df=pd.DataFrame(X)
sns.pairplot(df)
plt.show()

sns.set(style="darkgrid")

corr = df.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3,
            square=True, xticklabels=5, yticklabels=5,
            linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)
plt.show()
```

Proszę wygenerować próbkę z rozkładu Normalnego m = [0,0] oraz \Sigma = [[4.40, -2.75], [-2.75, 5.50]]

    Proszę policzyć średnią oraz macierz kowariancji próbki.
    Policz wektory i wartości własne macierzy kowariancji.
    Narysuj na jednym wykresie: próbkę poziomice rozkładu normalnego i wektory własne macierzy kowariancji przeskalowane przez wartości własne: 2*sqrt(eigenvalue)


```python
from scipy.stats import multivariate_normal

m = [0, 0]
Sigma = [[4.40, -2.75], [-2.75, 5.50]]
rv = multivariate_normal(m, Sigma)
data=rv.rvs(100)

x = data[:,0]
y = data[:,1]
mean = np.mean(data, axis=0)
cov = np.cov(x,y)

print("Srednią: ", mean)
print("Macierz kowariancji", cov)


def sampling(mean, cov, numb):
    x = np.random.multivariate_normal(mean, cov, numb)
    eigenvalues, eigenvectors = np.linalg.eig(cov)
    w, v = np.linalg.eig(cov)
    plt.scatter(x[:, 0],x[:, 1],c="grey")
    plt.scatter(mean[0],mean[1],c="red", alpha=1, s=20)
    
    plt.arrow(mean[0], mean[1], w[0]**0.5*v[0][0], w[0]**0.5*v[1][0], width = 0.1)
    plt.arrow(mean[0], mean[1], w[1]**0.5*v[0][1], w[1]**0.5*v[1][1], width = 0.1)
    plt.show()
    
numb = 1000
sampling(mean, cov, numb)
```
